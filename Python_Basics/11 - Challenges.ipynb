```python
def number_sum(n):
    sum = 0
    
    for i in range(1, n + 1):
        sum += i
    
    return sum

```

```python
def find_min(nums):
    min = float("inf")
    
    for num in nums:
        if num < min:
            min = num
    
    return min

```

```python
def remove_nonints(nums):
    new_nums = []
    
    for num in nums:
        if type(num) == int:
            new_nums.append(num)
    
    return new_nums

```

```python
def factorial(num):
    factorial = 1
    
    for i in range(1, num + 1):
        factorial *= i
    
    return factorial

```

```python
def area_sum(rectangles):
    sum = 0
    
    for rectangle in rectangles:
        sum += rectangle["height"] * rectangle["width"]
    
    return sum

```

```python
def divide_list(nums, divisor):
    new_list = []
    
    for num in nums:
        new_list.append(num / divisor)
    
    return new_list

```

```python
def join_strings(strings):
    joined = ""
    
    for s in strings:
        joined += s + ","
    
    if len(joined) != 0:
        joined = joined[:-1]
    
    return joined

```
